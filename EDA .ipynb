```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data 
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
df[['Age', "Orders","Amount"]].describe()
```

```python
#DATA Cleaning done: we have removed all the null values from the data 
```

```python
#EDA:
```

```python
#checking sales according to gender:
```

```python
ax=sns.countplot(x="Gender",data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=True)
ax=sns.barplot(x="Gender",y="Amount",data=sales)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#CONCLUSION-1: When it comes to both number and price, women have done more shopping than men.
```

```python
#Sorting on basis of age group
```

```python
df.columns
```

```python
ax=sns.countplot(data=df.sort_values(by='Age'),x="Age Group",hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
result=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=True)
ax=sns.barplot(x="Age Group",y="Amount",data=result)


sns.set(rc={'figure.figsize':(20,8)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#CONCLUSION-2: The age group of 26 to 35 has done the most shopping out of all the age groups. 
```

```python
df_grouped = df.groupby(['Age Group', 'Gender'])['Amount'].sum().reset_index()

g=sns.barplot(x='Age Group', y='Amount', hue='Gender', data=df_grouped)

for p in g.patches:
    g.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
#CONCLUSION-3: Women have shopped the most among those in the 26–35 age range.
```

```python
df.columns
```

```python
data_set=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=True)


ax=sns.barplot(x="State",y="Orders",data=data_set)
sns.set(rc={'figure.figsize':(20,15)})
```

```python
data_set=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=True)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x="State",y="Amount",data=data_set)
```

```python
#CONCLUSION-4: Among all the states, the most number of things ordered have come from Uttar Pradesh, Maharashtra, and Karnataka.
```

```python
df_grouped = df.groupby(['State', 'Gender'])['Amount'].sum().reset_index()

g=sns.barplot(x='State', y='Amount', hue='Gender', data=df_grouped)

for p in g.patches:
    g.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
data_set=df.groupby(["Marital_Status"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=True)
g=sns.barplot(x="Marital_Status",y="Amount",data=data_set)
sns.set(rc={'figure.figsize':(10,7)})


for p in g.patches:
    g.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
#CONCLUSION-5: Those who are not married have engaged in more shopping than those who are married.
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
g=sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')

for p in g.patches:
    g.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
```

```python
#on basis of zone:
data_set=df.groupby(["Zone","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=True)
g=sns.barplot(x="Zone",y="Amount",data=data_set,hue="Gender")
sns.set(rc={'figure.figsize':(23,17)})

for p in g.patches:
    g.annotate(f'{p.get_height():.4f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=15, color='blue', xytext=(0, 5), textcoords='offset points')
    
```

```python
#CONCLUSION-7: Out of all the zones, people from central zone have done more shopping.
```
